```python
import numpy as np
import pandas as pd
import tensorflow as tf
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('data.csv', sep=',', thousands=',',encoding='unicode_escape')
print(data.columns)
data
```

```python
def preprocess(data):
    max_cols = ['1fgpercx', '2fgpercx', '13percx', '23percx', '1record', '2record', '1fgperclg', '2fgperclg', '13perclg', '23perclg']
    for col in max_cols:
        max_val = max(data[col])
        data[col] /= max_val
    
    data['1recordx'] /= 5 #theoretically min is be 0
    data['2recordx'] /= 5 #theoretically min is be 0
    
    min_max_cols = ['1o/uline1', '1o/uline2','1o/uline3','1o/uline4','1o/uline5','2o/uline1','2o/uline2','2o/uline3',
                    '2o/uline4','2o/uline5', '1spreadx', '2spreadx', 'spread', 'o/u', '1ptsx', '1ptslg','2ptsx', 
                    '2ptslg','1ptsvar', '2ptsvar', '1astsx', '1astslg', '1rebsx', '1rebslg', '1pfsx', '1pfslg', 
                    '1tosx', '1toslg', '2astsx', '2astslg', '2rebsx', '2rebslg', '2pfsx', '2pfslg', '2tosx', '2toslg',
                    '1ftx','1ftlg', '2ftx', '2ftlg']
    for col in min_max_cols:
        max_val = max(data[col])
        min_val = min(data[col])
        data[col] = (data[col] - min_val)/max_val
    
    lst = []
    teams1 = {}
    counter = 0
    for i in range(len(data['team1'])):
        if '1'+data['team1'][i]  not in teams1.keys():
            teams1['1'+ data['team1'][i]] = counter
            lst.append(counter)
            counter += 1
        else:
            lst.append(teams1['1'+data['team1'][i]])
    data['team1'] = np.array(lst)
    
    lst = []
    teams2 = {}
    counter = 0
    for i in range(len(data['team2'])):
        if '2' + data['team2'][i]  not in teams2.keys():
            teams2['2'+data['team2'][i]] = counter
            lst.append(counter)
            counter += 1
        else:
            lst.append(teams2['2'+data['team2'][i]])
            
    data['team2'] = np.array(lst)
    
    team1 = pd.DataFrame(tf.one_hot(data.iloc[0:,0], 30).numpy())
    team2 = pd.DataFrame(tf.one_hot(data.iloc[0:,33], 30).numpy())
    team1.columns = teams1.keys()
    team2.columns = teams2.keys()
    
    data = data.join(team1)
    data = data.join(team2)
    
    cols = data.columns.to_list()
    cols = cols[0:1] + cols[33:34] + cols[1:33] + cols[34:64] + cols[67:] + cols[64:67]
    
    data = data[cols]
    return data
preprocess(data.copy())
```
