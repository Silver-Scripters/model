```python
import pandas as pd
import numpy as np
```

```python
lines = pd.read_csv('lines.csv', sep=',', thousands=',',encoding='unicode_escape')
```

```python
lines
```

```python
lines['point_total'] = lines['team1_score'] + lines['team2_score']
```

```python
lines['result'] = 'o'
for i in range(0, len(lines['result'])):
    if lines['point_total'][i] < lines['over_under'][i]:
         lines['result'][i] = 'u'
lines
```

```python
lines['point_diff'] = abs(lines['over_under'] - lines['point_total'])
for i in range(0, len(lines['point_diff'])):
    if lines['result'][i] == 'u':
        lines['point_diff'][i] *= -1
lines
```
